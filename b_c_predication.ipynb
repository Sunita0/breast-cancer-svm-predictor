```python
import pandas as pd 
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data, columns= data.feature_names)
df['target'] =data.target
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x= 'target',data= df)
plt.title('distribution 0= maligant,1= benign')
print(df['target'].value_counts())
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(df.corr(),cmap= 'coolwarm', annot= False)
plt.title('correlation heatmap of all features')
plt.show()
```

```python
feature_to_plot= ['mean radius', 'mean texture', 'mean area', 'mean concave points']
for feature in feature_to_plot:
    sns.histplot(data=df, x=feature, hue= 'target', kde= True, element='step')
    plt.title(f'Distribution of {feature} by class')
    plt.show()
```

```python
corr_target= df.corr()['target'].drop('target').sort_values(ascending= False) # +1 → strong positive relation with benign
#  -1 → strong negative relation (i.e. high values indicate malignancy)
print(corr_target)
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='target', y='mean radius', data=df)
plt.title('Mean Radius by Tumor Type')
plt.show()
plt.figure(figsize=(10, 5))
sns.boxplot(x='target', y='mean texture', data=df)
plt.title('Mean Texture by Tumor Type')
plt.show()
plt.figure(figsize=(10, 5))
sns.boxplot(x='target', y='mean area', data=df)
plt.title('Mean Area by Tumor Type')
plt.show()


```

```python
top_features= corr_target[abs(corr_target)>0.5]
print('top features correlation with target:\n',top_features)
```

# preprocessing

```python
#split features and target
x= df.drop('target', axis= 1)
y= df['target']
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=42, stratify=y)
# feature scaling
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x_train_scaled= scaler.fit_transform(x_train)
x_test_scaled= scaler.transform(x_test)
```

# training SVM model

```python
from sklearn.svm import SVC
svm_model= SVC(kernel='linear',random_state=42)
svm_model.fit(x_train_scaled,y_train)
y_pred= svm_model.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues',xticklabels=['Malignant','Benign'],yticklabels=['Malignant','Benign'])
plt.xlabel('predicted')
plt.ylabel('Actual')
plt.title('confusion matrix')
plt.show()    
```

```python
print(classification_report(y_test, y_pred, target_names=['Malignant', 'Benign']))
```

```python
import joblib

# Save model
joblib.dump(svm_model, 'svm_model.pkl')

# Save scaler
joblib.dump(scaler, 'scaler.pkl')
```
